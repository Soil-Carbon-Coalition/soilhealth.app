```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('C:/Users/Peter/Downloads/Fremont_Bridge_Hourly_Bicycle_Counts_by_Month_October_2012_to_present.csv', index_col='Date', parse_dates=True)
print(data.head())
```

```python
%matplotlib inline
data.resample('W').sum().plot()
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')
data.columns = ['West', 'East']
data.resample('W').sum().plot()
```

```python
#resample to daily with rolling sum (Note TWO sum() functions)

data['Total'] = data['West'] + data['East']
ax=data.resample('D').sum().rolling(365).sum().plot()
ax.set_ylim(0, None)
```

```python
data.groupby(data.index.time).mean().plot()
```

```python
pivoted = data.pivot_table('Total', index = data.index.time, columns = data.index.date)
pivoted.iloc[:5,:5]
```

```python
pivoted.plot(legend=False, alpha=0.01)
```
