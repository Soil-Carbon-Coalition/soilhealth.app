```python
startdate = "2000-01-01"
enddate = "2019-02-14"
# station_id = "375907091432201"
param_id = "72019" #well levels
data = DailyValueIO(
    start_date=startdate,
    end_date=enddate,
#     state='CO',
#     county='08003',
    basin="13010001",
#     station=station_id,
    parameter=param_id,
)
list(data)
```

```python
#make dataframe
levels = pd.DataFrame([[y for y in x] for x in data], columns=['name','code','variable_name','variable_code','unit','lat','lon','data'])
starts = []
ends = []
#add start and end dates for flow data
for site in data:
    starts.append(site.data[0][0].strftime('%Y-%m-%d')) #the first date in the tuple
    ends.append(site.data[-1][0].strftime('%Y-%m-%d')) #the last date in the tuple
levels['start'] = starts
levels['end'] = ends
```

```python
m = folium.Map(location=[40, -104.7], zoom_start=8, tiles="Stamen Terrain")
coords = [36,-121]
for i in range(0,len(levels)):
    folium.Marker([levels.iloc[i]['lat'], levels.iloc[i]['lon']],popup=levels.iloc[i]['start']+':'+levels.iloc[i]['end'],tooltip=levels.iloc[i]['name'], icon=folium.Icon(color='red')).add_to(m)
m
```
