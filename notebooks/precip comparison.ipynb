```python
# load some other data
hh = pd.read_excel('C:/Users/Peter/Documents/atlas/water/HH Rainfall 2010 to 2018.xlsx','Raw')
```

```python
#stack all the date and rain columns

x = [col for col in hh.columns if 'date' in col]
y = [col for col in hh.columns if 'rain' in col]

hh = pd.lreshape(hh, {'date': x,'rain': y})

def assign_wy(row):
    if row.date.month>=10:
        return(pd.datetime(row.date.year+1,1,1).year)
    else:
        return(pd.datetime(row.date.year,1,1).year)

hh['wy'] = hh.apply(lambda x: assign_wy(x), axis=1)

hh = hh.set_index(pd.DatetimeIndex(hh['date']))
hh = hh[hh['wy']==2012]

```
