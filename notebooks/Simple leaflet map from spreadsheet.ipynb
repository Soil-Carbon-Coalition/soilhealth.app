```python
import pandas as pd, numpy as np
import folium
import vincent, vega #for charting in popups
```

```python
df=pd.read_csv('C:/Users/Peter/Documents/atlas/atlasdata/Sites_all_Aug_2019.csv')
df
```

```python
m = folium.Map(location=[df['lat'].mean(), df['lon'].mean()], zoom_start=7, tiles="Stamen Terrain")
for i in range(0,len(df)):
    tooltip = df.iloc[i]['site']
    folium.Marker([df.iloc[i]['lat'], df.iloc[i]['lon']],tooltip=tooltip).add_to(m)
m
```
